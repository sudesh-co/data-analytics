```python

import  seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.get_dataset_names()
df= sns.load_dataset("iris")
df.head()
```

```python
df.species.value_counts().plot.barh()
```

```python
#Seaborn lmplot
sns.lmplot(
    data=df, x="sepal_length", y="sepal_width", col="species", hue="species",
    col_wrap=2, palette="muted", ci=1,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```

```python

sns.set_theme(style="darkgrid")

sns.displot(
    df, x="sepal_length", col="species",row=df["species"],
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
```

```python
sns.distplot(df["sepal_length"])
```

```python
#Kernal density estimations 

print(df.columns)
for  col in  df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        sns.kdeplot(df[col] , label=col)
    plt.legend()
```

```python
#Pairplot 
sns.pairplot(df, hue= 'species' ,  size=2.5)
```

```python
# Line plot 
new_df=df.head(200)
sns.lineplot(x= new_df["species"] , y=new_df["sepal_length"]  )
```
