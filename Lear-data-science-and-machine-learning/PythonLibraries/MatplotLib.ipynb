```python
#reading csv file
data = pd.read_csv('files/SuperStoreData.csv')
print(data.head())
# data.to_csv('files/SuperStoreData.csv', index=False)
```

```python
top_hundred_rec = data.head(10)
top_hundred_rec.plot(x='Sales' , y='Profit' , kind='bar')
```

```python
# Line 

top_hundred_rec = data.head(10)
top_hundred_rec.plot(x='Sales' , y='Profit' , kind='line')
```

```python
#area
top_hundred_rec = data.head(10)
top_hundred_rec.plot(x='Sales' , y='Profit' , kind='barh')
```

```python
#Pie Chart 
top_hundred_rec.plot(kind='pie', y='Sales', autopct='%1.1f%%')
```

```python
#Histogram 
top_hundred_rec.plot(x='Sales', y='Profit', kind='hist')
```

```python
#Area
top_hundred_rec.plot(x='Row ID', y='Discount', kind='area')
```
